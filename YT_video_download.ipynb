```python
import pandas as pd
import os
import youtube_dl
```

```python
input_file = r"D:\My Folder\Dailymotion\Videos.xlsx"
file = pd.read_excel(input_file)
file.astype('str').dtypes
```

```python
file.head()
```

```python
outtmpl = 'D:\My Folder\Dailymotion\Download_DM\script_download\%(title)s.%(ext)s'

ydl_opts = {
    'format': 'best',
    'outtmpl': outtmpl,
    'writethumbnail' : True,
    'embedthumbnail' : True
}

ydl = youtube_dl.YoutubeDL(ydl_opts)
```

```python
for i, row in file.iterrows():
    print("Trying to Download: ", row['yt_url'])
 
    try:
        d = ydl.download_with_info_file
        info = ydl.extract_info(row['yt_url'], download = True)
        
        download_message = 'Downloaded'
        
        file.loc[i, 'download_message'] =  download_message
        file.loc[i, 'yt_name'] = info['title']
        file.loc[i, 'yt_description'] = info['description']
        
        tags = ""
        for tag in info['tags']:
            tags = tags + tag + ","
        file.loc[i, 'yt_tags'] = tags
        
        file.loc[i, 'download_name'] = info['title'] + '.' + info['ext']
        file.loc[i, 'download_folder'] = 'D:\\My Folder\\Dailymotion\\Download_DM\\script_download'
        
        file.loc[i, 'thumbnail_name'] = info['title'] + '.jpg'
        
        print(download_message, " : ", info['title'], "\n\n")
        
    except Exception as e:
        download_message = 'Error in Download : ' + str(e)
        file.loc[i, 'download_message'] =  download_message
        print(download_message, "\n\n")
    
```

```python
file.to_excel(input_file)
```
