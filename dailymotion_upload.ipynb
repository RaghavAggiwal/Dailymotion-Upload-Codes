```python
import os
import dailymotion

d = dailymotion.Dailymotion()

_API_KEY_ENGLISH = "api_key"
_API_SECRET_ENGLISH = "api_Secret"
_USERNAME_ENGLISH = "x2b8iqv"
_PASSWORD_ENGLISH = "password"

_API_KEY_HINDI = "api_key"
_API_SECRET_HINDI = "api_secret"
_USERNAME_HINDI = "x2bgrnj"
_PASSWORD_HINDI = "password"
```

```python
upload_folder_english = r"D:\My Folder\Dailymotion\Download_DM\automated_upload"
upload_folder_hindi = r"D:\My Folder\Dailymotion\Download_DM\automated_upload_hindi"
```

```python
#uploading english videos

for file in os.listdir(upload_folder_english):
    try:
        
        print("Trying to upload: ", os.path.splitext(file))
        
        d.set_grant_type(
            "password",
            api_key=_API_KEY_ENGLISH,
            api_secret=_API_SECRET_ENGLISH,
            scope=["manage_videos"],
            info={"username": _USERNAME_ENGLISH, "password": _PASSWORD_ENGLISH},
        )
        
        # Uploading the file on dailymotion servers
        file_path = upload_folder_english +'\\' + os.path.splitext(file)[0] + ".mp4"
        url = d.upload(file_path)
        
        # Filling the information about the video
        parameters = {
            "url": url,
            "title": os.path.splitext(file)[0],
            "tags": "life,love,reality,god,spirituality,education,truth,saints,scriptures,meditation,religion,self,silence,\
            liberation,enlightenment,awakening,realisation,motivation,inspiration,anger,greed,hate,envy,jealousy,saints,\
            who am i,i am that,awareness,consciousness,nirvana,krishna,shiva,hinduism,veda,upanishad,\
            vedanta,mind,intelligence,acharya prashant,healing,yoga,mystic,happiness,joy,ego,spiritual retreat,yoga retreat",
            #"description": "my first automatic uplaod",
            "published": 1,
            "channel": "school"
        }

        # Sending the information to create the video on dailymotion
        result = d.post("/me/videos?fields=id,url", parameters)

        print("Uploaded video: ", os.path.splitext(file)[0], "\nDailymotion url: ", result['url'], "\n\n")
        

    except Exception as e:
        print("An error occured: %s" % str(e))
```
