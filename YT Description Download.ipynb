```python
import pandas as pd
import os
import youtube_dl
```

```python
input_file = r"D:\My Folder\Dailymotion\yt_description.xlsx"
file = pd.read_excel(input_file)
file.astype('str').dtypes
```

```python
file.head()
```

```python
outtmpl = 'D:\My Folder\Dailymotion\Download_DM\script_download_hindi\%(title)s.%(ext)s'

ydl_opts = {
    'format': 'best',
    'outtmpl': outtmpl,
    'writethumbnail' : True,
    'embedthumbnail' : True
}

ydl = youtube_dl.YoutubeDL(ydl_opts)
```

```python
for i, row in file.iterrows():
    print("Trying for: ", row['yt_url'])
 
    try:
        d = ydl.download_with_info_file
        info = ydl.extract_info(row['yt_url'], download = False)
        
        download_message = 'Done'
        
        description = "वीडियो जानकारी:\n"+info['description'].split('वीडियो जानकारी:',1)[1]
    
        
        file.loc[i, 'description'] =  description
        file.loc[i, 'name'] = info['title']
        file.loc[i, 'message'] = download_message
        
        print(download_message, "\n\n")
        
    except Exception as e:
        download_message = 'Error: ' + str(e)
        file.loc[i, 'download_message'] =  download_message
        print(download_message, "\n\n")
    
```

```python
file.to_excel(input_file)
```
